## 3. Vector Addition and Scalar Multiplication

```python
import numpy as np
import matplotlib.pyplot as plt
v = np.array([2, 3])
u = np.array([1, -1])

sum_vector = v + u
print("v + u =", sum_vector)
```

```python
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='v')
plt.quiver(v[0], v[1], u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b', label='u placed at end of v')
plt.quiver(0, 0, sum_vector[0], sum_vector[1], angles='xy', scale_units='xy', scale=1, color='g', label='v + u')

plt.xlim(-1, 5)
plt.ylim(-2, 5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.show()
```

```python
c = 0.5
scaled_v = c * v
print("2 * v =", scaled_v)

d = -1
scaled_v_neg = d * v
print("-1 * v =", scaled_v_neg)
```

```python
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='v')
plt.quiver(0, 0, scaled_v[0], scaled_v[1], angles='xy', scale_units='xy', scale=1, color='b', label='2 * v')
plt.quiver(0, 0, scaled_v_neg[0], scaled_v_neg[1], angles='xy', scale_units='xy', scale=1, color='g', label='-1 * v')

plt.xlim(-5, 5)
plt.ylim(-5, 7)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.show()
```

```python
combo = 3*v + (-2)*u
print("3*v - 2*u =", combo)
```

## 4. Linear Combinations and Span

```python
v = np.array([2, 1])
u = np.array([-1, 2])

combo1 = 2*v + 3*u
combo2 = -1*v + 4*u

print("2*v + 3*u =", combo1)
print("-v + 4*u =", combo2)
```

```python
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='v')
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b', label='u')
plt.quiver(0, 0, combo1[0], combo1[1], angles='xy', scale_units='xy', scale=1, color='g', label='2v + 3u')
plt.quiver(0, 0, combo2[0], combo2[1], angles='xy', scale_units='xy', scale=1, color='m', label='-v + 4u')

plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.show()
```

```python
# Generate many combinations
coeffs = range(-5, 6)
points = []
for a in coeffs:
  for b in coeffs:
    point = a*v + b*u
    points.append(point)

points = np.array(points)

plt.scatter(points[:, 0], points[:, 1], s=10, color='gray')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b')

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.show()
```

The gray dots show all reachable points with combinations of v and u.



```python
# dependent vectors


w = np.array([4, 2])
coeffs = range(-5, 6)
points = []
for a in coeffs:
  for b in coeffs:
    points.append(a*v + b*w)

points = np.array(points)

plt.scatter(points[:,0], points[:,1], s=10, color='gray')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='b')

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- 1. Definizione dei Vettori ---
f = np.array([2, 3, 4])
g = np.array([0, 1, 0])
h = np.array([1, 3, 0])

# --- 2. Calcolo della Combinazione Lineare ---
a, b, c = 5, -2, 3
combo = a * f + b * g + c * h

print("Vettore F:", f)
print("Vettore G:", g)
print("Vettore H:", h)
print(f"\nCombinazione calcolata ({a}*f + {b}*g + {c}*h): {combo}")

# --- 3. Setup per il Plot 3D ---
fig = plt.figure(figsize=(10, 8))
# Aggiunge gli assi 3D
ax = fig.add_subplot(111, projection='3d')

# Imposta i limiti degli assi per visualizzare bene tutti i vettori
max_coord = max(np.max(np.abs(f)), np.max(np.abs(g)), np.max(np.abs(h)), np.max(np.abs(combo)))
limit = int(max_coord) + 2 # Aggiunge un po' di margine
ax.set_xlim([-limit, limit])
ax.set_ylim([-limit, limit])
ax.set_zlim([-limit, limit])

# Etichette degli assi
ax.set_xlabel('Asse X')
ax.set_ylabel('Asse Y')
ax.set_zlabel('Asse Z')
ax.set_title('Combinazione Lineare di Vettori 3D')

# Funzione helper per plottare i vettori (quiver)
def plot_vector(vector, color, label):
    """Plotta un singolo vettore dall'origine (0,0,0)"""
    ax.quiver(0, 0, 0, vector[0], vector[1], vector[2],
              color=color, length=1, arrow_length_ratio=0.1, linewidth=2, label=label)

# --- 4. Plot dei Vettori Originali ---
plot_vector(f, 'r', 'f = [2, 3, 4] (Rosso)')
plot_vector(g, 'g', 'g = [0, 1, 0] (Verde)')
plot_vector(h, 'b', 'h = [1, 3, 0] (Blu)')

# --- 5. Plot del Vettore Combinazione (Risultato) ---
plot_vector(combo, 'k', f'Combo = {combo} (Nero, Risultato)',)

ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- 1. Definizione dei Vettori Generatori ---
# L'insieme {f, g, h}  linearmente indipendente, quindi il suo span  R^3.
f = np.array([2, 3, 4])  # Vettore con componente Z non nulla
g = np.array([0, 1, 0])  # Vettore sul piano XY
h = np.array([1, 3, 0])  # Vettore sul piano XY

# --- 2. Generazione dei Punti dello Span ---
# Usiamo coefficienti discreti nell'intervallo [-C, C] per visualizzare una "scatola" di punti.
C = 4
coeffs = np.arange(-C, C + 0.5, 0.5) # Vogliamo un po' pi di densit
points = []

# Tripletta di cicli per generare le combinazioni lineari
for a in coeffs:
    for b in coeffs:
        for c_val in coeffs:
            point = a * f + b * g + c_val * h
            points.append(point)

points = np.array(points)

# --- 3. Setup per il Plot 3D ---
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# --- 4. Plot dei Punti dello Span (Nuvola Grigia) ---
# Scatter plot dei punti generati. Questo  il campione visivo dello Span.
ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=5, color='gray', alpha=0.1, label='Span Sample (R)')

# --- 5. Plot dei Vettori Generatori ---
def plot_vector(vector, color, label):
    """Plotta un singolo vettore dall'origine (0,0,0)"""
    ax.quiver(0, 0, 0, vector[0], vector[1], vector[2],
              color=color, length=1, arrow_length_ratio=0.15, linewidth=2, label=label)

plot_vector(f, 'r', 'f = [2, 3, 4] (Rosso)')
plot_vector(g, 'g', 'g = [0, 1, 0] (Verde)')
plot_vector(h, 'b', 'h = [1, 3, 0] (Blu)')

# --- 6. Formattazione e Finalizzazione ---
limit = C * max(np.linalg.norm(v) for v in [f, g, h]) / 2
ax.set_xlim([-limit, limit])
ax.set_ylim([-limit, limit])
ax.set_zlim([-limit, limit])

ax.set_xlabel('Asse X')
ax.set_ylabel('Asse Y')
ax.set_zlabel('Asse Z')
ax.set_title('Span di Vettori 3D Indipendenti (Copertura di R)')

ax.legend()
plt.show()
```

## 5. Length (Norm) and Distance


```python
import numpy as np
import matplotlib.pyplot as plt
import math

# la norma / lunghezza di un vettore 2D si calcola con il teoremo di Pitagora

v = np.array([3, 4])
length = np.linalg.norm(v)
verbose_length = math.sqrt(v[0]**2 + v[1]**2)
print("Length of v =", length, verbose_length)
```

```python
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.text(v[0]/2, v[1]/2, f"Length={length}", fontsize=10, color='blue')
plt.grid()
plt.show()
```

Distance between two vectors The distance between v and another vector u is the length of their difference: v - u.

```python
u = np.array([0, 0])
dist = np.linalg.norm(v - u)
print("Distance between v and u =", dist)
```
